# II. Description des fichiers 'csv'

Programmation d'une boucle 'for', pour ressortir les données suivantes des fichiers 'brut'
- Le nombre de lignes
- Le nombre de colonnes
- Le nombre de ligne(s) en doublon(s)
- Le taux de remplissage pour chaque fichier

Si, présence d'une colonne vide, celle-ci est supprimée.

```python
# Liste pour stocker les 5 fichiers disponibles
files = ['EdStatsData.csv', 'EdStatsCountry.csv', 'EdStatsCountry-Series.csv', 'EdStatsFootNote.csv', 'EdStatsSeries.csv']

# Initialisation d'un dictionnaire pour stocker les informations
data_info = {}

# Boucle pour charger les fichiers et récupérer les informations
for file in files:
    # Charger le fichier CSV
    data = pd.read_csv(file)
    
# Supprimer les colonnes vides
    data = data.dropna(axis=1, how='all')
    
# Nombre de lignes en doublons dans les fichiers
    doublons = data.duplicated()
    
# Calculer le nombre de cellules vides
    cellules_vides = data.isnull().sum().sum()
    
# Stocker le nombre de lignes, de colonnes, de cellules vides et le taux de remplissage dans le dictionnaire
    data_info[file] = {
        'Nombre de lignes': data.shape[0],
        'Nombre de colonnes': data.shape[1],
        'Cellules vides': cellules_vides,
        'Nombres de doublons' : doublons.sum(),
        'Taux de remplissage': (data.size - cellules_vides) / data.size
    }

# Afficher les informations
for file, info in data_info.items():
    print(f"Informations pour le fichier {file}:")
    print(f"Nombre de lignes: {info['Nombre de lignes']}")
    print(f"Nombre de colonnes: {info['Nombre de colonnes']}")
    print(f"Nombre de cellules vides: {info['Cellules vides']}")
    print(f"Nombre de doublons: {info['Nombres de doublons']}")
    print(f"Taux de remplissage: {info['Taux de remplissage']:.2%}")
    print("\n")
```

**Description du contenu pour chaque fichier** : 

- __EdStatsData.csv__ :

Fichier qui compile l'ensemble des données liées directement aux nombreux KPI, à leur évoluation dans le temps, pour les différents pays du monde.

- __EdStatsCountry__ : 

Fichier qui compile tout une banque d'information pour les 241 pays, celui-ci peut servir de table de correspondance pour récupérer certaines données utiles pour approfondir ou compléter l'analyse. 

- __EdStatsCountrySeries__ : 

Fichier présentant quelques informations sur la source des données pour des indicateurs, selon les pays.

- __EdStatsFootNote__ : 

Fichier composé d'informations complémentaires, sur les KPI selon les pays et sur leur source de données.

- __EdStatsSeries__ :

Fichier qui compile une synthèse intéressante et exhaustive, de données complémentaires sur les KPI, ce fichier peut constituer une table de correspondance très utile.

Quand on regarde les différentes base de données, on remarque que 'EdStatsData.csv' présente toutes les données nécessaires au KPI et leur évolution dans le temps pour chaque pays. Par conséquent, elle est au coeur de notre analyse. Les autres sources de données sont des tables annexes qui peuvent nous servir si nous avons besoin d'information complémentaires à aggéger à notre table principale. Dans la partie suivante, nous allons nous intéresser à notre table principale et de son contenu, on remarque un faible taux de remplissage, nous allons donc devoir regarder plus en détails, pour la nettoyer et la structurer.

```python
# Filtrer mon dataframe data_main avec une liste d'indicateurs choisis au préalable, dans une liste
indicateurs_retenus = [
    "Internet users (per 100 people)",
    "Population, total",
    "GDP per capita (current US$)"
]
# Filtrer le DataFrame en fonction de la colonne "Indicator name"
data_main = data_main[data_main['Indicator Name'].isin(indicateurs_retenus)]
```

**B. Choix des années pertinentes pour l'étude**

Le choix des années à conserver est important dans le sens où la temporalité permet de déceler les évolutions des KPI à travers les années, par zone géographie ou pays. Pour la sélection des années, il y aura deux notions à prendre en compte. La première est la qualité des données en elles-mêmes et de leur exhaustivité par année, on regardera donc le taux de remplissage pour chacune d'entre elle. Puis, dans un second temps, il faudra mettre en perspective le besoin d'un point du vue fonctionnel, propre au notion de markéting.

**1. Calcul du taux de remplissage pour chacune des années**

```python
# Calcul du taux de remplissage par année, pour la base de données principale

# Condition si l'en-tête des colonnes est compris entre un interval déterminé
colonnes_a_analyser = [colonne for colonne in data_main.columns if 1970 <= pd.to_numeric(colonne, errors='coerce') <= 2100]

# Parcours des colonnes
for colonne in colonnes_a_analyser:
    
# Calcul du taux de remplissage pour chaque colonne
    taux_remplissage = 1 - data_main[colonne].isnull().sum() / len(data_main[colonne])
    
# Affichage du résultat
    print(f"Taux de remplissage de {colonne}: {taux_remplissage * 100:.2f}%")
```

**2. Représentation en histogramme avec 'matplotlib' du taux de remplissage selon les années**

```python
# Colonnes à analyser 
colonnes_a_analyser = [colonne for colonne in data_main.columns if 1970 <= pd.to_numeric(colonne, errors='coerce') <= 2016]

annees = []
taux_remplissage_list = []

for colonne in colonnes_a_analyser:
    # Calcul du taux de remplissage pour chaque colonne
    taux_remplissage = 1 - data_main[colonne].isnull().sum() / len(data_main[colonne])
    
    # Ajout des données à la liste
    annees.append(int(colonne))
    taux_remplissage_list.append(taux_remplissage * 100)

# Taille du graphique
plt.figure(figsize=(20, 10))

# Création du graphique en barres
bars = plt.bar(annees, taux_remplissage_list)

# Ajout des étiquettes de données sans chiffre après la virgule (arrondi)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{round(height)}%', ha='center', va='bottom')

plt.title('Taux de remplissage par année sur les 3 KPI retenus')
plt.xlabel('Années')
plt.ylabel('Taux de remplissage (%)')
plt.grid(axis='y')
plt.show()
```

Nous remarquons que selon les années, le taux de remplissage des colonnes n'est pas homogène. On peut en déduire que la concentration des données se situe à partir des années 2000 jusqu'en 2016, pour les 3 indicateurs.

**3. Conserver dans le DataFrame principal les années jugées pertinentes**

```python
# Dictionnaire avec les noms de colonnes à garder
annees_conservées = {'Country Name', 'Country Code','Indicator Code','Indicator Name','2000','2004','2008','2012','2016'}

# Obtenir les noms de colonnes communes
annees_conservées = list(annees_conservées.intersection(data_main.columns))

# Créer un nouveau DataFrame avec les colonnes années à conserver
data_main = data_main[annees_conservées]

# Organiser les colonnes selon l'ordre suivant
ordre_colonnes = ['Country Name', 'Country Code','Indicator Code','Indicator Name', '2000','2004','2008','2012','2016']
data_main = data_main[ordre_colonnes]
```

**4. Consolider l'ensemble des colonnes 'années' en une colonne unique**

```python
# Utilisation de la fonction 'pivot' pour regrouper les colonnes années en une colonne 'année'
data_main = data_main.melt(id_vars=['Country Name', 'Country Code','Indicator Code','Indicator Name'], var_name='Année', value_name='Valeur')
```

**5. Dépivoter la colonne 'Indicator Name' pour créer 3 colonnes distinctes pour chaque indicateur**

```python
data_main = data_main.pivot(index=['Country Name', 'Country Code', 'Année'], columns='Indicator Name', values='Valeur')
data_main = data_main.reset_index()
```

**6. Importation de deux colonnes supplémentaires d'une autre table**

```python
# Fusion des DataFrames en utilisant la colonne 'Country Code' comme clé de recoupement
data_main = pd.merge(data_main, data_EdStatsCountry[['Country Code', 'Income Group', 'Region']],
                       on='Country Code', how='left')

# Organiser les colonnes selon l'ordre suivant
ordre_des_colonnes = ['Region','Country Name', 'Income Group','Country Code','Année','Population, total','GDP per capita (current US$)','Internet users (per 100 people)']
data_main = data_main[ordre_des_colonnes]
```

**C. Analyse du taux de remplissage de chaque pays ou zones géographiques**

**D. Description et affichage du jeu de données avant analyse des indicateurs pour l'étude**

```python
# Création d'un dictionnaire pour stocker les informations
data_info_data_main = {}

# Nombre de lignes en doublons dans le fichier data_main
doublons_data_main = data_main.duplicated()

# Calculer le nombre de cellules vides dans le fichier data_main
cellules_vides_data_main = data_main.isnull().sum().sum()

# Stocker le nombre de lignes, de colonnes, de cellules vides et le taux de remplissage dans le dictionnaire
data_info_data_main['data_main'] = {
    'Nombre de lignes': data_main.shape[0],
    'Nombre de colonnes': data_main.shape[1],
    'Cellules vides': cellules_vides_data_main,
    'Nombres de doublons' : doublons_data_main.sum(),
    'Taux de remplissage': (data_main.size - cellules_vides_data_main) / data_main.size
}

# Afficher les informations pour le fichier data_main
print("Informations pour le fichier data_main:")
print(f"Nombre de lignes: {data_info_data_main['data_main']['Nombre de lignes']}")
print(f"Nombre de colonnes: {data_info_data_main['data_main']['Nombre de colonnes']}")
print(f"Nombre de cellules vides: {data_info_data_main['data_main']['Cellules vides']}")
print(f"Nombre de doublons: {data_info_data_main['data_main']['Nombres de doublons']}")
print(f"Taux de remplissage: {data_info_data_main['data_main']['Taux de remplissage']:.2%}")
```

**Calcul du coéfficient de corrélation entre les variables économique & numérique**

```python
# Copie du dataframe principal
data_coef = data_main_final.copy()

# Convertir la colonne 'année' en type datetime
data_coef['Année'] = pd.to_datetime(data_coef['Année'], format='%Y')

# Initialisation d'une liste 
correlation_coefficients = []

# Itération sur les années
for year in data_coef['Année'].unique():
    
    df_year = data_coef[data_coef['Année'] == year]
    
    # Calculer le coefficient de corrélation pour cette année spécifique
    correlation = df_year['Internet users (per 100 people)'].corr(df_year['GDP per capita (current US$)'])
    
    # Ajouter le coefficient de corrélation et l'année correspondante à la liste
    correlation_coefficients.append({'Year': year, 'Correlation': correlation})

# Conversion de liste en dataframe
correlation_df = pd.DataFrame(correlation_coefficients)

# Afficher le dataframe
correlation_df.head()
```

# VIII.Analyse des trois variables grâce à un nuage de points

**Nuage de points selon le classement de développement des pays 'Income Group' en  2016**

```python
#Groupby sur les colonnes spécifiées
resultats_groupby = data_filtre.groupby('Income Group').agg({
    'Population, total': 'sum',
    'GDP per capita (current US$)': 'mean',
    'Internet users (per 100 people)': 'mean'
}).reset_index()

# Calcul des moyennes des abscisses et des ordonnées
moyenne_x = resultats_groupby['Internet users (per 100 people)'].mean()
moyenne_y = resultats_groupby['GDP per capita (current US$)'].mean()

# Création du nuage de points
plt.figure(figsize=(22, 18))
scatterplot = sns.scatterplot(
    x='Internet users (per 100 people)',
    y='GDP per capita (current US$)',
    size='Population, total',
    data=resultats_groupby,
    hue='Income Group',
    sizes=(500, 10000),
    palette='viridis', 
    legend='full'
)

# Ajout de légendes et de titres
plt.title('Nuage de points par Income Group')
plt.xlabel('Moyenne Internet users (per 100 people)')
plt.ylabel('Moyenne GDP per capita (current US$)')

# Ajout d'étiquettes de région à chaque point
for line in range(0, resultats_groupby.shape[0]):
     scatterplot.text(resultats_groupby['Internet users (per 100 people)'][line],
                      resultats_groupby['GDP per capita (current US$)'][line],
                      resultats_groupby['Income Group'][line],
                      horizontalalignment='left',
                      size='medium',
                      color='black',
                      weight='semibold'
                     )

# Ajout des lignes de moyennes
plt.axvline(x=moyenne_x, color='red', linestyle='--', label='Moyenne Internet users')
plt.axhline(y=moyenne_y, color='blue', linestyle='--', label='Moyenne GDP per capita')

# Déplacement de la légende dans le coin supérieur gauche et agrandissement du cadre
plt.legend(loc='upper left', bbox_to_anchor=(4, 4), borderaxespad=0.)

# Affichage du graphique
plt.legend()
plt.show()
```

# IX. Etablissement d'un 'Score' pour classer les pays

**Classement pour la population totale en 2016**

```python
# Sélectionner les colonnes 'Country Name', 'Population, total' et 'Année'
colonnes_selectionnees = ['Country Name', 'Population, total', 'Année']
nouveau_dataframe = data_main[colonnes_selectionnees]

# Filtrer le DataFrame pour l'année 2016
nouveau_dataframe_2016 = nouveau_dataframe[nouveau_dataframe['Année'] == 2016]

# Trier le dataframe en fonction de la colonne 'Population, total' dans l'ordre décroissant
classement_population = nouveau_dataframe_2016.sort_values(by='Population, total', ascending=False)

# Réinitialiser l'index pour obtenir une colonne d'index numérique et créer une nouvelle colonne 'Numero'
classement_population = classement_population.reset_index(drop=True)
classement_population['Classement Population'] = classement_population.index + 1
```

**Classement pour le PIB par habitant en 2016**

```python
# Sélectionner les colonnes 'Country Name', 'Population, total' et 'Année'
colonnes_selectionnees = ['Country Name', 'GDP per capita (current US$)', 'Année']
nouveau_dataframe = data_main[colonnes_selectionnees]

# Filtrer le DataFrame pour l'année 2016
nouveau_dataframe_2016 = nouveau_dataframe[nouveau_dataframe['Année'] == 2016]

# Trier le dataframe en fonction de la colonne 'Population, total' dans l'ordre décroissant
classement_GDP = nouveau_dataframe_2016.sort_values(by='GDP per capita (current US$)', ascending=False)

# Réinitialiser l'index pour obtenir une colonne d'index numérique et créer une nouvelle colonne 'Numero'
classement_GDP = classement_GDP.reset_index(drop=True)
classement_GDP['Classement Economique'] = classement_GDP.index + 1
```

**Classement pour l'accès à internet en 2016**

```python
# Filtrer le dataframe pour exclure les pays de la liste
data_main = data_main[~data_main['Country Name'].isin(pays_a_exclure)]

# Sélectionner les colonnes 'Country Name', 'Population, total' et 'Année'
colonnes_selectionnees = ['Country Name', 'Internet users (per 100 people)', 'Année']
nouveau_dataframe = data_main[colonnes_selectionnees]

# Filtrer le DataFrame pour l'année 2016
nouveau_dataframe_2016 = nouveau_dataframe[nouveau_dataframe['Année'] == 2016]

# Trier le dataframe en fonction de la colonne 'Population, total' dans l'ordre décroissant
classement_internet = nouveau_dataframe_2016.sort_values(by='Internet users (per 100 people)', ascending=False)

# Réinitialiser l'index pour obtenir une colonne d'index numérique et créer une nouvelle colonne 'Numero'
classement_internet = classement_internet.reset_index(drop=True)
classement_internet['Classement Numérique'] = classement_internet.index + 1
```

**Classement globalisant les trois indicateurs pour chaque pays**

```python
classement_global = pd.merge(classement_population, classement_GDP[['Country Name', 'Classement Economique']], on='Country Name', how='inner')
classement_global = pd.merge(classement_global, classement_internet[['Country Name','Classement Numérique']], on='Country Name', how='inner')
classement_global['Résultat global'] = classement_global['Classement Population']+classement_global['Classement Economique']+classement_global['Classement Numérique']
classement_global= classement_global.sort_values(by='Résultat global', ascending=True)
classement_global = classement_global.drop(['Population, total', 'Année'], axis=1)
```
