# Transformation de la table annexe 'EdStatsCountry'

**Importation de la librairie 'Pandas' & du fichier 'EdStatsCountry'**

```python
#Importation de la librairie Pandas et OS
import pandas as pd
import os
```

```python
#Importation du fichier en format 'csv' --> 'EdStatsCountry'
data_country = pd.read_csv('EdStatsCountry.csv')
```

**Remplacement de la valeur 'Central Asia & Europe' dans la colonne région par 'Europe' & 'Asie Centrale' selon les pays**


```python
# Remplacement des valeurs dans la table 'EdStatsCountry' par la création de liste pour créer des tables de correspondance
pays_central_asia = [
    'Armenia', 
    'Azerbaijan',
    'Belarus',
    'Georgia',
    'Kazakhstan',
    'Kyrgyz Republic',
    'Russia',
    'Tajikistan',
    'Turkey',
    'Turkmenistan',
    'Ukraine',
    'Uzbekistan'
]

pays_europe = [
    'Albania', 
    'Andorra',
    'Austria',
    'Belgium',
    'Bosnia and Herzegovina',
    'Bulgaria',
    'Channel Islands',
    'Croatia',
    'Cyprus',
    'Czech Republic',
    'Denmark',
    'Estonia',
    'Faeroe Islands',
    'Finland',
    'France',
    'Germany',
    'Gibraltar',
    'Greece',
    'Greenland',
    'Hungary',
    'Iceland',
    'Ireland',
    'Isle of Man',
    'Italy',
    'Kosovo',
    'Latvia',
    'Liechtenstein',
    'Lithuania',
    'Luxembourg',
    'Macedonia',
    'Moldova',
    'Monaco',
    'Montenegro',
    'Netherlands',
    'Norway',
    'Poland',
    'Portugal',
    'Romania',
    'San Marino',
    'Serbia',
    'Slovak Republic',
    'Slovenia',
    'Spain',
    'Sweden',
    'Switzerland',
    'United Kingdom'
]

pays_south_america = [
    'Argentina',
    'Bolivia',
    'Brazil',
    'Chile',
    'Colombia',
    'Ecuador',
    'Guyana',
    'Paraguay',
    'Peru',
    'Suriname',
    'Uruguay',
    'Venezuela'
]

pays_central_america = [
    'Belize',
    'Costa Rica',
    'El Salvador',
    'Guatemala',
    'Honduras',
    'Mexico',
    'Nicaragua',
    'Panama'
]

pays_caribbean = [
    'Antigua and Barbuda',
    'Aruba',
    'Barbados',
    'British Virgin Islands',
    'Cayman Islands',
    'Cuba',
    'Curaçao',
    'Dominica',
    'Dominican Republic',
    'Grenada',
    'Haiti',
    'Jamaica',
    'Puerto Rico',
    'Sint Maarten (Dutch part)',
    'St. Kitts and Nevis',
    'St. Lucia',
    'St. Martin (French part)',
    'St. Vincent and the Grenadines',
    'The Bahamas',
    'Trinidad and Tobago',
    'Turks and Caicos Islands',
    'Virgin Islands'
]

pays_east_asia_pacific = [
    'Nauru'
]

def update_region(row):
    if row['Short Name'] in pays_central_asia:
        return 'Central Asia'
    elif row['Short Name'] in pays_europe:
        return 'Europe'
    elif row['Short Name'] in pays_south_america:
        return 'South America'
    elif row['Short Name'] in pays_central_america:
        return 'Central America'
    elif row['Short Name'] in pays_caribbean:
        return 'Caribbean' 
    elif row['Short Name'] in pays_east_asia_pacific:
        return 'East Asia & Pacific'
    else:
        return row['Region']

# Appliquer la fonction update à la colonne 'Region'
data_country['Region'] = data_country.apply(update_region, axis=1)
```

**Création d'un nouveau fichier csv avec le dataframe 'data_country'**

```python
# Nom du nouveau fichier
nom_fichier_csv = 'EdStatsCountryNew.csv'

# Obtenir le chemin d'accès
repertoire_courant = os.getcwd()

# Chemin vers le répertoire pour stocker le fichier
EdStatsCountryNew_csv = os.path.join(repertoire_courant, nom_fichier_csv)

# Enregistrement du dataframe dans un nouveau fichier CSV
data_country.to_csv(EdStatsCountryNew_csv, index=False)
```
